 ## Loops
In each of next 2 cells you will have to perform operations with loops.

```python
# From n_list get elements with modulo 0 and 2 when divided by 3 and store to modulo_02 in form of a list.
n_list = [2, 3, 6, 7, 8, 9, 10, 11, 17]
modulo_02 = []

for num in n_list:
    if (num % 3 == 0 or num % 3 == 2):
        modulo_02.append(num)
modulo_02
```

```python
# Create list called fibonacci_75 with first 75 elements of Fibonacci sequence.
# You have here an algoritm that will print you first the 75 elements,
# but you need to store it to list, not only print.
n1 = 0
n2 = 1
count = 0
fibonacci_75 = [n1, n2]

while count < 75:
    nth = n1 + n2
    fibonacci_75.append(nth)
    n1 = n2
    n2 = nth
    count += 1
fibonacci_75
```
